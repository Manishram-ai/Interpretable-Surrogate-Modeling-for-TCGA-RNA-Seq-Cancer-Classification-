## Load Phenotype data

```r
file_path <- "/content/TCGA_phenotype_denseDataOnlyDownload.tsv"


pheno_data <- read.table(
  file_path,
  header = TRUE,          # The first row is a header
  sep = "\t",             # The file is tab-separate          # The first column contains row names
)
```

```r
pheno_data <- pheno_data[pheno_data$sample_type == "Primary Tumor" | pheno_data$sample_type == "Solid Tissue Normal", ]
```

```r
data_gene = data$sample
```

## Transpose the data of RNA Seq

```r
transpose = t(data)
```

```r
dim(transpose)
```

```r
transpose
```

```r
tumor_status_df <- data.frame(
  tumor_status = pheno_data$sample_type,
  sample = pheno_data$sample,
  cancer_type = pheno_data$X_primary_disease
)
```

```r
transpose_df <- as.data.frame(transpose)
```

## Check the dimention and NaN values

```r
total_rows <- nrow(transpose_df)

# Count the number of rows that have at least one NA value
rows_to_remove <- sum(!complete.cases(transpose_df))

# Print the results
cat(sprintf("Total number of rows in the data frame: %d\n", total_rows))
cat(sprintf("Number of rows that will be removed (contain NA): %d\n", rows_to_remove))
```

```r
# transpose and column names
# combined data
```

```r
dim(column_names_df)
```

```r
dim(transpose_df)
```

```r
full_df <- cbind(column_names_df,transpose_df)
```

```r
# Extract current names from full_df
current_names <- colnames(full_df)
current_names
```

```r
# Extract new names from data_gene
# Assuming data_gene has them in a column named 'gene' (change if different)

new_gene_names <- data_gene
```

```r
# Replace all except the first ("sample")
colnames(full_df) <- c(current_names[1], new_gene_names)
```

```r
full_df
```

## Merged Based on sample

```r
combined_data <- merge(full_df, tumor_status_df, by = "sample")
```

```r
dim(combined_data)
```

```r
total_rows <- nrow(combined_data)

# Count the number of rows that have at least one NA value
rows_to_remove <- sum(!complete.cases(combined_data))

# Print the results
cat(sprintf("Total number of rows in the data frame: %d\n", total_rows))
cat(sprintf("Number of rows that will be removed (contain NA): %d\n", rows_to_remove))
```

## Download the data as csv file

```r
write.csv(combined_data, "final_rna_combined_data_with_column_names.csv", row.names = FALSE)
```
